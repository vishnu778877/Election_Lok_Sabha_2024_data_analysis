# Analysis of Lok Sabha Election Result 2024

## Importing Python Libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing Dataset

```python
 
df = pd.read_excel('election_result_2024.xlsx')
df.head()
```

## Checking any Null Values

##### To thoroughly inspect dataset for any missing values, we'll check each column to ensure data integrity. Identifying and handling null values is a crucial step in the data cleaning process.


```python
df.isnull().sum()
```

## Key Insights from dataset:

##### We are gathering key insights from the dataset, such as the total number of constituencies in the Lok Sabha election, details of seats won by the BJP, INC, and other parties, the top 10 parties in terms of winning seats, and ## the distribution of vote share percentage. These insights help us understand the overall election results and the performance of major political parties.

## Total Number of Constituencies in Lok Sabha election 2024

```python
total_constituencies = df['constituency_name'].nunique()
print("Total constituencies in Lok Sabha election 2024:", total_constituencies)
```

## Details of seats won by Bhartiya Janata Party (BJP)

```python
df.loc[(df['party_name'] == 'Bharatiya Janata Party') & (df['rank'] == 1)]
```

#### Based on the analysis of the election data using the loc method, we observe that the Bharatiya Janata Party (BJP) has won a total of 240 seats.

## Details of seats won by Indian National Congress (INC)

```python
df.loc[(df['party_name'] == 'Indian National Congress') & (df['rank'] == 1)]
```

### Based on the analysis of the election data using the loc method, we observe that the Indian National Congress(INC) has won a total of 99 seats.

## Details of seats won by Others

```python
df.loc[(df['party_name'] != 'Indian National Congress') & (df['party_name'] != 'Bharatiya Janata Party')& (df['rank'] == 1)]
```

## Top 10 Party in Number of seats won 

```python
winners = df[df['rank'] == 1]
seats_won = winners['party_name'].value_counts()
top_10_parties = seats_won.head(10)
seats_won_final = top_10_parties

plt.figure(figsize=(12, 8))
ax = seats_won_final.plot(kind='barh', color='blue')
plt.title('Number of Seats Won by Each Party (Top 10)')
plt.xlabel('Number of Seats')
plt.ylabel('Party')
plt.show()
```

## Votes Share % by Party

```python
total_votes_by_party = df.groupby('party_name')['total_votes'].sum()
top_10_votes = total_votes_by_party.nlargest(10)


plt.figure(figsize=(12, 8))
top_10_votes.plot(kind='pie', autopct='%1.1f%%', title='Votes % Distribution by Party')
plt.ylabel('')
plt.show()
```
